## 1 产品物理特性

<a id='v'></a>
### 平均电压
[return](#c)

根据图像算法组提供的电压数据每小时计算一次这一小时内所有电压值的均值，可以分析出电压随时间的变化情况，阈值上界为13V，下界为11V。其计算公式为：
$$\bar{V}=\sum_{i=1}^{n}V_i$$
其中$V_i$是任意这一小时读取到的电压值。

<a id='v1'></a>
### 电压稳定度
[return](#c)

每小时计算一次以该时间段内电压值为样本的极差，用以描述电压的波动情况，阈值待定。其计算公式为
$$\max{V_i}-\min{V_j}$$
其中$V_i$,$V_j$为一小时内读取到的任意电压值。

<a id='c1'></a>
### 平均电流
[return](#c)

每小时计算一次电流的均值，分析电流随时间的变化情况，阈值待定。其计算公式为：
$$\bar{I}=\sum_{i=1}^{n}I_i$$
其中$I_i$是任意这一小时读取到的电流值。

<a id='c2'></a>
### 电流稳定度
[return](#c)

每小时计算一次电流数据的极差，分析电流的波动情况，阈值待定。其计算公式为：
$$\max{I_i}-\min{I_i}$$
其中$I_i$为一小时内读取到的电流值。

<a id='d1'></a>
### PWM值
[return](#c)

取每小时的平均值，用来衡量电机寿命。其计算公式为：
$$\bar{P}=\sum_{i=1}^{n}P_i$$
其中$P_i$是任意这一小时读取到的PWM值。

<a id='d2'></a>
### 帧用时分布区间占比
[return](#c)

该指标反映一小时内的每帧用时在其均值周围的分布情况。设标准为$D=66.66$ms，将分布区间分为$[D-0.01,D+0.01]$, $[D-0.02,D-0.01)\cup(D+0.01,D+0.02]$, $[D-0.03,D-0.02)\cup(D+0.02,D+0.03]$, $(-\infty,D-0.03)\cup(D+0.03,+\infty)$, 统计落在以上各个区间的测量值所占的比例。以下两类图分别展示整体和个体的情况。

<a id='d3'></a>
### 平均每帧用时
[return](#c)

该指标考察电机转动的稳定性，其计算公式为：
$$\frac{\sum_{i=1}^NT_i}{N}$$
其中$T_i$为每小时数据文件中的每帧用时，$N$为该文件中的总帧数。

## 2 产品传输接口特性
<a id='z1'></a>
### CRC校验错误率
[return](#c)

统计每个小时的数据文件中未能通过CRC校验的帧定义为错误帧，判据原理计算公式：  
$$\frac{错误总帧数}{总帧数}\times100\%\leq阈值$$

<a id='z2'></a>
### 不完整帧占比
[return](#c)

将整帧数据按空格分割后长度不对（不等于124）的帧定义为不完整帧，该判据计算公式为：
$$\frac{不完整的总帧数}{总帧数}\times100\%\leq阈值$$

<a id='z3'></a>
### 漏帧率
[return](#c)

考察样机传输数据的稳定性。判据原理公式：
$$\frac{N_1}{N_2}\times100\%\leq 0.2\%$$
其中$N_1$表示当前小时漏帧总个数, $N_2$表示当前小时总帧数, 0.2%是设定的阈值(第一次寿命试验中的阈值)。

## 3 数据统计特性
<a id='s1'></a>
### 小时平均值 
[return](#c)

用来监控测量值的稳定性和趋势性，其计算公式为：
$$\bar{X_h}=\frac{1}{n}\sum_{i=1}^n X_i$$
其中$X_i$为每小时内每帧某个角度的测量值。

### 小时最大值
用来监控测量值的稳定性和趋势性，其计算公式为：
$$ {\rm Max}_h=\max X_i$$
其中$X_i$为每小时内每帧某个角度的测量值。
### 小时最小值
用来监控测量值的稳定性和趋势性，其计算公式为：
$$ {\rm Min}_h=\min X_i$$
其中$X_i$为每小时内每帧某个角度的测量值。

<a id='s11'></a>
### 测量值日平均值
[return](#c)

用来监控以日为单位的测量值的稳定性和趋势性，其计算公式为：
$$\bar{X_d}=\frac{1}{N}\sum_{i=1}^N {\bar X}_i$$
其中${\bar X}_i$为第i个小时某个角度的测量均值，$N$为该日测得的小时均值的个数。
### 测量值日最大值
用来监控以日为单位的测量值的稳定性和趋势性，其计算公式为：
$${\rm Max}_d=\max {\rm Max}_{h_i}$$
其中${\rm Max}_{h_i}$为第i个小时某个角度的测量值最大值。
### 测量值日最小值
用来监控以日为单位的测量值的稳定性和趋势性，其计算公式为：
$${\rm Min}_d=\max {\rm Min}_{h_i}$$
其中${\rm Min}_{h_i}$为第i个小时某个角度的测量值最小值。

<a id='s2'></a>
### 测量值随机误差最大值
[return](#c)

该判据提取了去除异常值后的数据各个角度中与数据平均值之差的最大值，研究传感器的随机误差的大小及稳定性。已标定的设备以SICK为标准，阈值为30mm。其判据计算公式：
$$\max_i(X_i-\bar X_h)$$
其中$X_i$为一个小时内一个位置的任意测量值，$\bar X_h$表示该小时内该位置的平均值。

<a id='s3'></a>
### 日波动最大值
[return](#c)

每小时测量数据的最值与日均值的差，阈值为30mm。计算公式如下：
$$\max\left\{\left|\max{X_i}-\bar{X_{d}}\right|,\left|\min{X_i}-\bar{X_{d}}\right|\right\}$$
其中$X_i$为该小时内任意测量值，$\bar{X}_{d}$为每日整体均值

<a id='s4'></a>
### 测量值标准差
[return](#c)

该指标可以监控测量值的波动情况，阈值待定。其计算公式如下所示：
$$s=\sqrt{\frac{1}{n}\sum_{i=1}^n\left(X_i-\bar X_h\right)^2}$$
其中$X_i$为一个小时内某个位置的任意测量值，$\bar X_h$为该小时内对应位置的均值，$n$为该位置测量值的个数。

<a id='s5'></a>
### 测量值极差
[return](#c)

该指标用来监控测量值的波动情况，阈值为60mm，其计算公式如下：
$$\max X_i-\min X_j$$
其中$X_i$和$X_j$是某一位置的任何两个测量值，该指标即该位置的最大值与最小值的差

<a id='s6'></a>
### 测量值分布区间占比
[return](#c)

该指标反映一小时内的各测量值在均值周围的分布情况。设均值为$\mu$，将分布区间分为$[\mu-10,\mu+10]$, $[\mu-20,\mu-10)\cup(\mu+10,\mu+20]$, $[\mu-30,\mu-20)\cup(\mu+20,\mu+30]$, $(-\infty,\mu-30)\cup(\mu+30,+\infty)$, 统计落在以上各个区间的测量值的比例。

<a id='s7'></a>
### 随机误差最值出现的角度频率
[return](#c)

统计每台设备每个小时的测量数据中随机误差最大值出现的角度信息，每个小时文件一个角度值。

<a id='s8'></a>
### 小时平均值与真实值之差
[return](#c)

取试验环境稳定状态下设备运行几个小时后的数据的均值作为初始状态，监控样机长时间运行后与初始状态的变化，即系统误差的变化趋势。判据计算公式如下：
$$X_h-X_{true}$$
其中$X_h$表示一个小时内某位置输出值的平均值，$X_{true}$表示对应位置的真实值。

<a id='s9'></a>
### 日均值波动
[return](#c)

每小时测量数据的均值与每日整体均值的差，阈值为10mm。计算公式如下：
$$\bar{X}_{h}-\bar{X}_{d}$$
其中$\bar{X}_{h}$为每小时的均值，$\bar{X}_{d}$为每日整体均值。

<a id='s10'></a>
### 产品误差区间
[return](#c)

在已知真实值的情况下，求每小时测量数据的误差范围，阈值为$\pm$20mm，计算公式如下：
$$[\min{X_i}-X_{true},\max{X_i}-X_{true}]$$
其中$X_i$为这一小时内某一角度的任意测量值，$X_{true}为对应角度的真实值$。
